```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('BankRecords.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.drop(['ID'], axis=1)
```

# EDA
## 1. Target Variable Distribution

```python
plt.figure(figsize=(8, 4))
plt.hist(df["Income(Thousands's)"], bins=25, edgecolor='k', alpha=0.7)
plt.title('Distribution of Yearly Income')
plt.xlabel('Income (Thousands)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

## 2. Numerical Variables Distribution

```python
numerical_variables = ['Age','Experience(Years)', 'Sort Code','Family','Credit Score',"Mortgage(Thousands's)"]
```

```python
num_vars = len(numerical_variables)

fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# flattening axes
axes = axes.flatten()

# Plot histograms for each numerical variable
for i, var in enumerate(numerical_variables):
    axes[i].hist(df[var], bins=30, edgecolor='k')
    axes[i].set_title(f'Distribution of {var}')
    axes[i].set_xlabel(var)
    axes[i].set_ylabel('Frequency')
    axes[i].grid(True)


plt.tight_layout()
plt.show()
```

## 3. Categorical Variables Distribution

```python
categorical_variables = ['Education','Personal Loan','Securities Account','CD Account','Online Banking','CreditCard']
```

```python
# getting numerical variables length
num_vars = len(categorical_variables)

# Creating subplots
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# flatenning the axes
axes = axes.flatten()

# barplots
for i, var in enumerate(categorical_variables):
    counts = df[var].value_counts()
    axes[i].bar(counts.index, counts.values, color='skyblue', edgecolor='k', alpha=0.7)
    axes[i].set_title(f'Distribution of {var}')
    axes[i].set_xlabel(var)
    axes[i].set_ylabel('Frequency')
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```

```python
# creating subplots
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# flatenning the axes
axes = axes.flatten()

# boxplots
for i, var in enumerate(categorical_variables):
    df.boxplot(column="Income(Thousands's)", by=var, ax=axes[i], grid=False)
    axes[i].set_title(f'Income Distribution by {var}')
    axes[i].set_xlabel(var)
    axes[i].set_ylabel('Income (Thousands)')
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=45)

plt.tight_layout()
plt.show()
```

## 4. Checking Linearity in numerical variables

```python
# getting correlation matrix
corr_matrix = df[numerical_variables].corr()

# plotting heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

## 4. Scaling Techniques

### Log

```python
# applying log transformation
df['Log_Income(Thousands)'] = np.log1p(df["Income(Thousands's)"])

# plotting distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Log_Income(Thousands)'], bins=30, edgecolor='k', alpha=0.7)
plt.title('Distribution of Log-Transformed Income (Thousands)')
plt.xlabel('Log of Income (Thousands)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### min-max

```python
from sklearn.preprocessing import StandardScaler

# initialising standard scaler
scaler = StandardScaler()

# creating scaled variables
df['Scaled Income(Thousands)'] = scaler.fit_transform(df[["Income(Thousands's)"]])

# plotting scaled variable
plt.figure(figsize=(10, 6))
plt.hist(df['Scaled Income(Thousands)'], bins=25, edgecolor='k', alpha=0.7)
plt.title('Distribution of scaled Income (Thousands)')
plt.xlabel('Min-Max Normalized Income (Thousands)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
